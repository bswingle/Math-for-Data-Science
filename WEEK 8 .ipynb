```python
A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
              [np.sqrt(2)/2, np.sqrt(2)/2]])
```

```python
x = np.array([[2], [-2]])
```

```python
T = np.array([[2.82842712], [0.]])
```

```python
trans = T + np.dot(A, x)
xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
xx, yy = np.meshgrid(xx, yy)
xy = np.vstack([xx.ravel(), yy.ravel()])
```

```python
trans_grid = T + np.dot(A, xy)
```

```python
xx_transformed = trans_grid[0].reshape(xx.shape)
yy_transformed = trans_grid[1].reshape(yy.shape)
```

```python
f, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[0].set_title('Original Points')
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
axes[1].set_title('Transformed Points')
plt.show()
```
