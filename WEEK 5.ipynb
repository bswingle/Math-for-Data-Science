```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def right_riemann_sum(f, a, b, n):
    delta_x = (b-a) / n
    x_values = np.linspace(a, b, n+1)
    y_values = f(x_values[1:])
    sum_ = np.sum(y_values * delta_x)
    return sum_, x_values[1:], y_values
```
