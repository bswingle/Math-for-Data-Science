```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import openpyxl
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
file_path = r'C:\Users\icyde\OneDrive - Bellevue University\YEAR 2\SEMESTER 1\MATH\WEEK 3\us_pop_data.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
print("Original Data:")
print(data.head())
```

```python
data['Years_Since_1790'] = data['year'] - 1790
data['Population_in_Millions'] = data['us_pop'] / 1000000
print(data[['Years_Since_1790', 'Population_in_Millions']])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Years_Since_1790'], data['Population_in_Millions'], marker='o', linestyle='-')
plt.title('US Population Over Time')
plt.xlabel('Years_Since_1790')
plt.ylabel('Population_in_Millions)')
plt.grid(True)
plt.show()
```

```python
X = data[['Years_Since_1790']]
y = data['Population_in_Millions']
model = LinearRegression()
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_
predicted_population = model.predict(X)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', marker='o', alpha=0.5)
plt.plot(X, slope * X + intercept, color='red', linewidth=2)  # <-- Corrected line
plt.title('US Population Over Time')
plt.xlabel('Years_Since_1790')
plt.ylabel('Population_in_Millions')
plt.grid(True)
plt.show()
```

```python
r_squared = r2_score(y, predicted_population)
print("R-squared value:", r_squared)
```

```python
data['Years_Since_1790'] = data['year'] - 1790
data['Population_in_Millions'] = data['us_pop'] / 1000000
data['Years_Since_1790_Squared'] = data['Years_Since_1790'] ** 2
print(data[['Years_Since_1790', 'Population_in_Millions', 'Years_Since_1790_Squared']])
```

```python
X_squared = data[['Years_Since_1790_Squared']]
y = data['Population_in_Millions']
model_squared = LinearRegression()
model_squared.fit(X_squared, y)
predicted_population = model_squared.predict(X_squared)
slope = model_squared.coef_[0]
intercept = model_squared.intercept_
plt.figure(figsize=(10, 6))
plt.scatter(X_squared, y, color='blue', marker='o', alpha=0.5)
plt.title('US Population Over Time')
plt.xlabel('Years_Since_1790_Squared')
plt.ylabel('Population_in_Millions')
plt.grid(True)
plt.plot(X_squared, slope * X_squared + intercept, color='red', linewidth=2)
plt.show()
```

```python
r_squared = r2_score(y, predicted_population)
print("R-squared value:", r_squared)
```
