```python
import numpy as np
from numpy.linalg import eig
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
A = np.array([[4, 0, 1],
              [-1, -6, -2],
              [5, 0, 0]])
```

```python
v = np.array([[1],
             [2],
             [3]])
```

```python
Av = np.dot(A, v)
```

```python
is_eigenvector = np.allclose(Av, 2 * v)
```

```python
if is_eigenvector:
    print("The vector [2] is an eigenvector of matrix A.")
else:
    print("The vector [2] is not an eigenvector of matrix A.")
```

```python
file_path = r'C:\Users\icyde\OneDrive - Bellevue University\YEAR 2\SEMESTER 1\MATH\WEEK 9\video_game_data.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
user_scores = data['user_score']
```

```python
critic_scores = data['critic_score']
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(range(len(user_scores)), user_scores, color='blue', marker='o', alpha=0.5, label='User Scores')
plt.scatter(range(len(critic_scores)), critic_scores, color='red', marker='o', alpha=0.5, label='Critic Scores')
plt.xlabel('Critic Scores')
plt.ylabel('User Scores')
plt.title('User Scores vs. Critic Scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X = data[['critic_score', 'user_score']]
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
pc1 = pca.components_[0]
pc2 = pca.components_[1]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(range(len(user_scores)), user_scores, color='blue', marker='o', alpha=0.5, label='User Scores')
plt.scatter(range(len(critic_scores)), critic_scores, color='red', marker='o', alpha=0.5, label='Critic Scores')
plt.arrow(pca.mean_[0], pca.mean_[1], pc1[0]*3, pc1[1]*3, color='purple', width=0.05)
plt.arrow(pca.mean_[0], pca.mean_[1], pc2[0]*3, pc2[1]*3, color='green', width=0.05)
plt.xlabel('Critic Scores')
plt.ylabel('User Scores')
plt.title('User Scores vs. Critic Scores with Principal Components')
plt.grid(True)
plt.show()
```
